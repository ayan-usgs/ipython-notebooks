```python
import netCDF4 as nc4
from pysgrid import from_nc_dataset  # check object type and to create sgrid object rather than separate methods
```

```python
DATASET_URL = 'http://geoport.whoi.edu/thredds/dodsC/coawst_4/use/fmrc/coawst_4_use_best.ncd'
ds = nc4.Dataset(DATASET_URL)
```

```python
sg = from_nc_dataset(ds)
```

```python
u = ds.variables['u']
u_data = u[1, 1, u_var.center_slicing[2], u_var.center_slicing[3]]
```

```python
u_data.shape
```

```python
from pysgrid.processing_2d import avg_to_cell_center
u_avg_cell_center = avg_to_cell_center(u_data, u_var.center_axis)
```

```python
u_avg_cell_center.shape
```

```python
v_var = sg.v
v = ds.variables['v']
v_data = v[1, 1, v_var.center_slicing[2], v_var.center_slicing[3]]
v_data.shape
```

```python
v_avg_cell_center = avg_to_cell_center(v_data, v_var.center_axis)
v_avg_cell_center.shape
```

```python
from pysgrid.processing_2d import rotate_vectors, vector_sum
angles = ds.variables['angle'][sg.angle.center_slicing]
u_rot, v_rot = rotate_vectors(u_avg_cell_center, v_avg_cell_center, angles)  # need method to calculated angles from lat/lon for delft3d and wrf... # kyle has a function he likes
```

```python
uv_vector_sum = vector_sum(u_rot, v_rot)
```
